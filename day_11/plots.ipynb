```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Importing dataset
df = sns.load_dataset("titanic")

df
```

```python
df.info()
```

```python
# Check null values
df.isnull().sum()
```

```python
sns.countplot(x="survived",hue="sex",data =df)
```

```python
# Scatter plot
plt.scatter(x="age",y='fare', data=df)
```

```python
sns.scatterplot(x="age",y='fare', data=df, hue="pclass")
```

```python
sns.barplot(x="sex",y='fare', data=df, hue="pclass")
```

```python
sns.pairplot(df)
```

```python
df.hist(bins=10,grid=False,figsize=(9,7))
```

```python
sns.boxplot(x="sex",y='fare',hue="who", data=df)
```

```python
sns.boxenplot(x="sex",y='fare',hue="who", data=df)
```

```python
f= sns.FacetGrid(df, hue="survived", col="pclass",margin_titles=True,
                palette = {1: "seagreen", 0: "gray"})
f= f.map(plt.scatter, "fare", "age",edgecolor="g").add_legend()
```

```python
f= sns.FacetGrid(df, hue="survived", col="sex",margin_titles=True,
                palette = {1: "seagreen", 0: "gray"})
f= f.map(plt.scatter, "fare", "age",edgecolor="g").add_legend()
```

```python
sns.stripplot(x="sex",y='fare',hue="who", data=df)
```

```python
# Heat map
numeric_columns = df.select_dtypes(include=['number'])
corr = numeric_columns.corr()

plt.figure(figsize=(10, 10))
sns.heatmap(corr)
```

```python
sns.violinplot(x="sex",y='fare',hue="who", data=df)
```
