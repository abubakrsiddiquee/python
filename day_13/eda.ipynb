# Exploratory Data Analysis
Exploratory Data Analysis is a method used to `analyze` and `summerize` dataset. In Python, you can use the `Pandas` library to perform EDA on dataset.





```python
# Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import dataset
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
# Summerize (Summary Statistics)
df.describe()
```

```python
# Find Null Values
df.isnull().sum()
```

```python
df.isnull().sum()*100/len(df)
```

```python
# Finding Unique Values
df.nunique()
```

```python
df.columns
```

```python
# Visualize the NAN values in heatmap
sns.heatmap(df.isnull(), yticklabels=False , cbar=False , cmap='viridis')
```

```python
df.drop(columns="deck" , inplace=True)
```

```python
df.columns
```

```python
# Visualize and Explore
sns.countplot(data= df ,x= "survived", hue="survived" ,legend=False)
```

```python
sns.barplot(data=df, x="survived", y="fare", hue="who")
```

```python
sns.countplot(data=df, x="pclass" ,hue="pclass", legend=False)
```

```python
# Data Distributtion
sns.histplot(data=df, x='age', bins=30, color="orange")
```

```python
df['age'].hist(bins= 30, color= "yellow",alpha=0.7)
```

```python
sns.boxplot(data=df, y="age", x="survived" , hue="who")
```

```python
sns.scatterplot(x="age", y="fare", data=df, hue="pclass")
plt.ylim(0,300)
```

```python
sns.pairplot(data=df, hue="survived")
```

```python
#stackedbar plot
df.groupby(["pclass", "who"])['fare'].mean().unstack().plot(kind='bar', stacked= True)
```

```python
#Swarm Plot
sns.swarmplot(data=df, x='survived',y='age', hue='who')
```

```python
sns.catplot(x='pclass',y='fare', hue='embarked',kind='point',data=df)
```

```python
#Relationship of two numeric variable
filtered_df = df[df['fare'] < 300]
sns.lmplot(data=filtered_df,x='age', y='fare',hue='pclass')
```

```python
#How to add a column and make age group
df.columns
```

```python
# Adding Column
df['age_groups']= pd.cut(df['age'],[0,10,18,30,50,80],labels=['Kids','Child','Youth','Middle Age', 'Old']) 
```

```python
df.head()
```

```python
df.groupby('age_groups')['fare'].mean().plot(kind='bar')
```

```python
sns.pairplot(df,hue='survived', vars=['fare','pclass','age'])
```
